```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r"/Users/sa8/Desktop/Titanic/csv_data/train.csv"


train_data = pd.read_csv(file_path)


df = train_data
```

```python
df
```

```python
df_train= df.copy()
```

```python
# Handling the missing data


# Fill missing 'Age' values with the median age based on passenger class ('Pclass') and gender ('Sex')
df_train['Age'] = df.groupby(['Pclass', 'Sex'])['Age'].transform(lambda x: x.fillna(x.median()))

# Create a new column 'HasCabin', where 1 means the passenger has a cabin, and 0 means no cabin info
df_train['HasCabin'] = df['Cabin'].apply(lambda x: 0 if pd.isnull(x) else 1)

# Fill missing 'Embarked' values with the most common port (mode) where passengers boarded
most_common_embarked = df_train['Embarked'].mode()[0] if not df_train['Embarked'].mode().empty else 'S'  # 'S' is a common default
df_train['Embarked'].fillna(most_common_embarked, inplace=True)

# Convert 'Sex' column to numeric (0 for male, 1 for female)
df_train['Sex'] = df_train['Sex'].map({'male': 0, 'female': 1})
```

```python
#Dropped columns with objects in them and dropped the columns that didn't seem useful
df_train = df_train.drop(columns=['Cabin','Embarked','Name','PassengerId','Ticket'])

df_train 
```
